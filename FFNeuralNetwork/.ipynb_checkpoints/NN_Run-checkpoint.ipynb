```python
def closest (y_pred, classes):
    new_y_pred = y_pred.copy()
    for i in range(len(y_pred)):
        curr = classes[0]
        for c in classes:
            if abs(y_pred[i] - c) < abs(y_pred[i] - curr):
                curr = c

        y_pred[i] = curr
    return y_pred
```
